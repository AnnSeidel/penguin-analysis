```python
def parseFile(infile):
    with open(infile, 'r') as infile:
        for line in infile:
            pass
        total,common,perc=line.split('\t')
    return int(total),int(common),float(perc)

def adjust_labels(v2, B_x, B_y, left_x, left_y,right_x, right_y):
    lbl = v2.get_label_by_id("B")
    x, y = lbl.get_position()
    lbl.set_position((x+B_x, y+B_y))

    lbl = v2.get_label_by_id("10")
    x, y = lbl.get_position()
    lbl.set_position((x+left_x, y+left_y))

    lbl = v2.get_label_by_id("01")
    x, y = lbl.get_position()
    lbl.set_position((x+right_x, y+right_y))

def plotVenn(in1, in2, title=False):

    for key in label_dic:
        if os.path.basename(in1).startswith(key):
            TOOL1=key
        if os.path.basename(in2).startswith(key):
            TOOL2=key

    A_total,AB,_= parseFile(in1)
    B_total,BA,_= parseFile(in2)

    labelsExt=['%s (%d)' %(label_dic[TOOL1],A_total), '%s (%d)' %(label_dic[TOOL2],B_total)]
    outfile=os.path.splitext(sys.argv[1])[0] + ".png"

    A_only=A_total-AB
    B_only=B_total-BA

    v2=venn2(subsets=(A_only,B_only,AB),\
             set_labels = (label_dic[TOOL1],label_dic[TOOL2]),\
             set_colors=(color_dic[TOOL1], color_dic[TOOL2]), alpha=0.8);
    venn2_circles(subsets=(A_only,B_only,AB), linewidth=0.5)# linestyle="--");
    if title:
      plt.title("Overlap of complete phage genomes")
    
    v2.get_label_by_id('01').set_text('%.2f%%   %.2f%%' %(BA/B_total*100.00, B_only/B_total*100.00))
    
    v2.get_label_by_id('10').set_text('%.2f%%   %.2f%%' %(A_only/A_total*100.00,AB/A_total*100.00))
    v2.get_label_by_id('11').set_text('')
   
    return v2;

def plotVennWithoutLabels(in1, in2, title=False):

    figureSize = (2.88/2.54, 2.5/2.54)
    for key in label_dic:
        if os.path.basename(in1).startswith(key):
            TOOL1=key
        if os.path.basename(in2).startswith(key):
            TOOL2=key

    A_total,AB,_= parseFile(in1)
    B_total,BA,_= parseFile(in2)

    labelsExt=['%s (%d)' %(label_dic[TOOL1],A_total), '%s (%d)' %(label_dic[TOOL2],B_total)]
    outfile=os.path.splitext(sys.argv[1])[0] + ".png"

    A_only=A_total-AB
    B_only=B_total-BA

    v2=venn2(subsets=(A_only,B_only,AB),\
             set_labels = ("",""),#(label_dic[TOOL1],label_dic[TOOL2]),\
             set_colors=(color_dic[TOOL1], color_dic[TOOL2]), alpha=0.8);
    #venn2_circles(subsets=(A_only,B_only,AB), linewidth=0.5)# linestyle="--");

    
    for idx, subset in enumerate(v2.subset_labels):
        v2.subset_labels[idx].set_visible(False)
    plt.gcf().set_size_inches(figureSize)
    plt.gcf().tight_layout()
    #fig.tight_layout()
    #v2.get_label_by_id('01').set_text('%.2f%%   %.2f%%' %(BA/B_total*100.00, B_only/B_total*100.00))
    
    #v2.get_label_by_id('10').set_text('%.2f%%   %.2f%%' %(A_only/A_total*100.00,AB/A_total*100.00))
    #v2.get_label_by_id('11').set_text('')
   
    return v2;
```

```python
import plotly.graph_objects as go

from itertools import combinations

tool="penguin"
trust_order_count=0
columns=["protId", "startCoord", "endCoord", "Strand", "partial", "contigId", "contigLen", "protein", "score"]
protein_file = os.path.join(summaryDir, "proteins", "{}.complete_phage.proteins.tsv".format(tool))
df_proteins = pd.read_csv(protein_file, sep='\t', names=columns)
df_proteins=df_proteins.loc[df_proteins["partial"] == 0]
tmp = df_proteins.loc[df_proteins["Strand"] == -1]["startCoord"]
df_proteins.loc[df_proteins["Strand"] == -1,"startCoord"] = df_proteins.loc[df_proteins["Strand"] == -1]["contigLen"] - df_proteins.loc[df_proteins["Strand"] == -1]["endCoord"] + 1
df_proteins.loc[df_proteins["Strand"] == -1,"endCoord"] = df_proteins.loc[df_proteins["Strand"] == -1]["contigLen"] - tmp + 1
df_proteins["Strand"] = 1
df_proteins=df_proteins.sort_values(["contigId","startCoord","endCoord"])

core_proteins=(df_proteins["protein"].unique())
cooccurence=np.zeros((len(core_proteins),len(core_proteins)), dtype=int)
index={"MP_A":0, "MP_B":1, "MP_C":2,  "CP_A":3, "CP_B":4, "CP_C":5, "CP_D":6, "CP_E":7, "CP_F":8,"CP_G":9,"CP_H":10,"RdRp_B":11, "RdRp_A":12}
countProteins={}

for contigId in df_proteins["contigId"].unique():
  #if contigId == "SRR7976310_60885":
  #  continue
  order=df_proteins.loc[df_proteins["contigId"] == contigId,"protein"].tolist()
  if len(order) in countProteins:
    countProteins[len(order)]+=1
  else:
    countProteins[len(order)]=1


  order = list(set(order))
  order.sort(key=lambda x:index[x])
  for i in range(len(order)-1):
    #if index[order[i]] != index[order[i+1]]:
        cooccurence[index[order[i]]][index[order[i+1]]] += 1 

source=[]
target=[]
value=[]
for i in range(len(cooccurence)):
  for j in range(len(cooccurence[i])):
    if cooccurence[i][j] > 0:
      source.append(i)
      target.append(j)
      value.append(cooccurence[i][j])
link = dict(source=source, target=target, value=value)
node=dict(label=list(index.keys()), thickness=15, pad=15)
print(link)
data=go.Sankey(link=link, node=node)
fig=go.Figure(data)
import plotly.io as pio
w=2204
h=720
width_in_mm=100
#scale = (width_in_mm / 25.4) / (w / 300)
scale=1
pio.write_image(fig, 'image.png', width=w, height=h, scale=scale)
config = {
  'toImageButtonOptions': {
    'format': 'png', # one of png, svg, jpeg, webp
    'filename': 'custom_image',
    'height': h,
    'width': w,
    'scale':scale # Multiply title/legend/axis/canvas sizes by this factor
  }
}


fig.show(config=config)
```
