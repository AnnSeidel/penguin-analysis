```python
import sys, os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
label_dic={"penguin":"PenguiN",
           "megahit":"Megahit",
           "metaspades":"metaSPAdes",
           "metaviralspades":"metaviralSPAdes",
           "rnaspades":"rnaSPAdes",
           "rnaviralspades": "rnaviralSPAdes",
           "iva":"IVA",
           "vicuna": "VICUNA",
           "savage": "SAVAGE",
           "haploflow":"Haploflow"
          }

color_dic={"penguin":'#EE8866',
           "megahit":'#EEDD88',
           "metaspades":'#77AADD',
           "metaviralspades":'#FFAABB',
           "rnaspades":'#99DDFF',
           "rnaviralspades":'#44BB99',
           "iva":'#D6E465',
           "vicuna": '#DDDDDD',
           "savage": '#a38995',
           "haploflow": '#AAAA00'
        }


color2Label={label_dic[label]:color_dic[label] for label in label_dic}
```

```python
plt.rcParams["font.size"] = 6
DPI=300
```

```python
outDir="../figures/"

if not os.path.exists(outDir):
    os.mkdir(outDir)
```

# Figure 7

```python


def numfmt(x, pos):
    if x>0:
        s = '{:d}k'.format(int(x // 1000))
    else:
        s = '{:d}'.format(int(x))
    return s

import matplotlib.ticker as tkr
xfmt = tkr.FuncFormatter(numfmt)
```

```python
# Figure 7 A
figureSize=(5.6/2.54, 4.0/2.54)
def plot_horizontal_50perccov_99clust(ax):
    assemblers0 = ("penguin", "megahit", "metaspades", "rnaspades",
     "rnaviralspades", "iva", "vicuna", "haploflow")
    assemblers = [label_dic[key] for key in assemblers0]
    rrnas_counts = (113196, 9123, 1713, 18933, 3203, 229, 2028, 1059)

    ax.barh(assemblers, rrnas_counts, align='center',
     color=[color_dic[key] for key in assemblers0])
    ax.invert_yaxis()
    ax.set_xlabel('Non-redundant\nassembled 16S rRNA contigs')
    #ax.tick_params(axis='both', which='major', labelsize=8)
    #ax.tick_params(axis='both', which='minor', labelsize=8)
    #ax.patch.set_edgecolor('black')
    #ax.patch.set_linewidth(1)

    #ax.xaxis.set_ticks(np.array([0, 20000, 50000, 70000, 100000]))
    ax.set_xlim(0, 120000)
fig=plt.figure()
#Set size
plot_horizontal_50perccov_99clust(plt.gca())
plt.gca().xaxis.set_major_formatter(xfmt)
fig.set_size_inches(figureSize)
fig.tight_layout()
plt.savefig(outDir + "Fig_7_A.png", dpi=DPI, transparent=True)
```
